- Muchas variables tienen valores minimos de 0, o como la columna 2 que que el minimo es negativo. Podrian deberse a datos ausentes.

### Asimetria y curtosis:

- Muchas variables tienen curtosis menor a -1, lo que indica distribuciones mas planas, es decir, con menos valores extremos.
